## DIG for time series data

```python
import numpy as np
from DIG import DIG
import matplotlib.pyplot as plt
import scipy as sci
from scipy import io
import phate
```

```python
# Load EEG data 
data = io.loadmat(r'C:\Users\Andres Duque\Documents\Research2\n10_8-12hz.mat')
labels_d = io.loadmat(r'C:\Users\Andres Duque\Documents\Research2\labelsn10.mat')
labels = labels_d['hyp']
eeg = np.transpose(data['data'])
labels  = np.reshape(labels, 860)
```

```python
# select some channels 
Data_eeg = eeg[:,5:10]
```

```python
# Compute histograms distances using DIG
dig = DIG(L1=3840, L3=3840)
hist = dig.compute_hist_distances(Data_eeg)
```

```python
# Compute PHATE using the histograms distances 
phate_emb = phate.PHATE(knn_dist='precomputed_distance', knn = 20)
phate_fit = phate_emb.fit_transform(hist.histogram_distances)
```

```python
import scprep
scprep.plot.scatter2d(phate_fit, c = labels, title = "Sleep stages", legend_loc = "upper left")
```

# Influence of the number of bins in the histograms and the covariance time window





```python
Number_bins = np.array([10, 20, 50])
L2_window = np.array([5, 20, 50])
```

```python
embeddings = []
for i in Number_bins:
    for j in L2_window:
        dig = DIG(L1=3840, L3=3840, n_bins = i, L2 = j)
        hist = dig.compute_hist_distances(Data_eeg)
        phate_emb = phate.PHATE(knn_dist='precomputed_distance', knn = 20)
        phate_fit = phate_emb.fit_transform(hist.histogram_distances)
        title = "n_bins = " + str(i) + " L2 = " + str(j)
        embedding = [phate_fit, title]
        embeddings.append(embedding)
```

```python

for i in range(len(embeddings)):
    title = embeddings[i][1]
    scprep.plot.scatter2d(embeddings[i][0], c = labels, title = "Sleep stages, " + title, legend_loc = "upper left")
```

# Euclidean distances 

```python
Number_bins = np.array([10, 20, 30, 40, 50])
embeddings = []
for i in Number_bins:
    dig = DIG(L1=3840, L3=3840, n_bins = i, histograms_distance = 'euclidean')
    hist = dig.compute_hist_distances(Data_eeg)
    phate_emb = phate.PHATE(knn_dist='precomputed_distance', knn = 20)
    phate_fit = phate_emb.fit_transform(hist.histogram_distances)
    title = "n_bins = " + str(i) 
    embedding = [phate_fit, title]
    embeddings.append(embedding)
```

```python
for i in range(len(embeddings)):
    title = embeddings[i][1]
    scprep.plot.scatter2d(embeddings[i][0], c = labels, title = "Sleep stages E\n uclidean distance in the histograms space" + title, legend_loc = "upper left")
```
